```python
import os
import shutil
import sys

from google.colab import drive
# Force remount to refresh credentials
drive.mount('/content/drive', force_remount=True)
# Define source folder
sys.path.append(os.path.abspath('/content/drive/MyDrive/Colab Notebooks/scheduling/xNet/'))
source_folder = "/content/drive/MyDrive/Colab Notebooks/scheduling/xNet/test1"  # Main folder where files & common folders are
destination_base = "test"  # Base name for subfolders
suffix = "_xNet"  # Suffix for subfolder names

# Define common folders & file that need to be copied to each subfolder
common_folders = ["statistics", "routings", "graphs", "__pycache__"]  # Adjust folder names
common_file = "datanetAPI.api"  # Adjust the file name

# Get a sorted list of all files in the source folder (excluding common folders/files)
files = sorted([f for f in os.listdir(source_folder) if f.startswith("gnnet_data_set")])

# Ensure we have exactly 100 files to distribute
if len(files) < 20:
    raise ValueError("Not enough unique files in the source folder (needs 100).")

# Loop to create subfolders and move files
for i in range(1, 19):
    subfolder_name = f"{destination_base}{i}{suffix}"
    subfolder_path = os.path.join(source_folder, subfolder_name)

    # Create subfolder if it doesn’t exist
    os.makedirs(subfolder_path, exist_ok=True)

    # Move one unique file into the new subfolder
    file_to_move = files[i - 1]  # Take one file at a time
    shutil.move(os.path.join(source_folder, file_to_move), os.path.join(subfolder_path, file_to_move))

    # Copy common folders into each subfolder
    for folder in common_folders:
        src_folder_path = os.path.join(source_folder, folder)
        dest_folder_path = os.path.join(subfolder_path, folder)
        if os.path.exists(src_folder_path):
            shutil.copytree(src_folder_path, dest_folder_path, dirs_exist_ok=True)

    # Copy the common file into each subfolder
    src_file_path = os.path.join(source_folder, common_file)
    dest_file_path = os.path.join(subfolder_path, common_file)
    if os.path.exists(src_file_path):
        shutil.copy2(src_file_path, dest_file_path)

    print(f"✅ Created {subfolder_name}, moved {file_to_move}, and copied common files/folders.")

print("🎉 Task completed: 100 subfolders created with files and common items.")
```
